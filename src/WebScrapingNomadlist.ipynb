```python
def buscar_celda_valor(indice_valor, tabla_detalles):
    celda = None
    
    indice = 0
    for tabla in tabla_detalles:
        values = tabla.find_all("td", class_ = "value")
        #display(values)
        for value in values:
            if indice == indice_valor:
                celda = value.contents[0]
                #print("Celda encontrada - Indice:", indice, " - celda:", celda)
                break
                
            indice += 1

        if celda != None:
            break
            
    return celda

def get_valor_internet(celda_internet):
    '''
    <div class="rating r5">
        <div class="filling">
            <span class="hide-before-800px">🏎 Fast: </span>34Mbps<span class="hide-before-800px"> (avg)</span>
        </div>
    </div>
    https://www.crummy.com/software/BeautifulSoup/bs4/doc/#tag
    '''
    #display(type(celda_internet))
    #display(list(celda_internet.children))
    return celda_internet.findChildren("div" , recursive=False)[0].contents[1]

def get_valor_ranking(celda_ranking):
    '''
    <div class="rating r4">
        <div class="filling" style="width:82.402510686395%">
            <span xitemprop="ratingValue">4.12</span>/<span xitemprop="bestRating">5</span> (Rank #37)
        </div>
    </div>
    '''
    # 4.12 ....contents[0].contents[0]
    # 5 ....contents[2].contents[0]
    return 0
```

```python
req = urllib.request.Request(
    url, 
    data=None, 
    headers={
        'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_9_3) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/35.0.1916.47 Safari/537.36'
    }
)
html = urllib.request.urlopen(req)
sopa = BeautifulSoup(html)
tabla_detalles = sopa.find_all("table", class_ = "details")
#display(tabla_detalles)
display(type(tabla_detalles))
display(type(tabla_detalles[0]))
print("Numero de tablas:",len(tabla_detalles))
valores = {}
indice = 0
for tabla in tabla_detalles:
    keys = tabla.find_all("td", class_ = "key")
    #print(keys)
    for key in keys:
        clave = key.contents[0][2:].strip()
        print(key, key.contents[0][:2].strip(), clave, sep='|')
        valores[clave]= {"indice": indice, "valor": ""}
        indice += 1
display(valores)

'''
for tabla in tabla_detalles:
    values = tabla.find_all("td", class_ = "value")
    #print(keys)
    for value in values:
        clave = key.contents[0][2:].strip()
        print(key, key.contents[0][:2].strip(), clave, sep='|')
        valores[clave]= {"indice": indice, valor: ""}
        indice += 1
'''
indice_internet = 4
celda_internet = buscar_celda_valor(indice_internet, tabla_detalles)
print("Resultado buscar_celda_valor:", celda_internet)
get_valor_internet(celda_internet)
```
