<a id="variables"></a>
 
### Descripción de las variables

- **propertyCode**: código únido que identifica la vivienda
- **address**: dirección de la vivianda
- **province**: provincia en la que se encuantra la vivienda
- **municipality**: municipio en el que se encuantra la vivienda
- **district**: distrito en el que se encuentra la vivienda
- **country**: código del país en el que se encuentra la vivienda  
- **neighborhood**: barrio en el que se encuentra la vivienda
- **latitude** y **longitude**: coordenada geográfica de la vivienda
- **distance**: distancia al centro en metros
- **floor**: planta
- **price**: precio de alquiler al mes
- **propertyType**: tipo de propiedad. Posibles valores: 'flat', 'chalet', 'duplex', 'penthouse', 'studio', 'countryHouse'
- **size**: tamaño de la vivienda en m2
- **exterior**: indica su es un piso exterior (True) o interior (False)
- **rooms**: número de habitaciones de la vivienda
- **bathrooms**: número de cuartos de baño de la vivienda
- **status**: estado de la vivienda. Posibles valores: 'good', 'renew', 'newdevelopment'
- **hasLift**: indica si tiene ascensor. 1.0 tiene ascensor, 0.0 no tiene ascensor

### Tratamiento de Missings

```python
idealista_fe.nan_analysis(idealista_df);
```

Para la prueba de concepto del desafío no vamos a obtener el barrio y el distrito de las viviendas que no la tienen, directamente eliminamos los registros sin barrio o distrito. Para un proyecto real habría que recurrir a otras fuentes para obtener el barrio y el distrito.

```python
idealista_df = idealista_df[idealista_df['neighborhood'].notna() & idealista_df['district'].notna()]
idealista_fe.nan_analysis(idealista_df);
```

```python
idealista_df.loc[idealista_df['hasLift'].isna(),'propertyType'].value_counts()
```

Para los chalet y casas de campo asumimos que no tienen ascensor, y eliminamos el resto de viviendas con *hasLift* a nan. Para un proyecto real habría que recurrir a otras fuentes para determinar si esas viviendas tienen o no ascensor.

```python
idealista_df.loc[idealista_df['hasLift'].isna() & ((idealista_df['propertyType']=='chalet') | (idealista_df['propertyType']=='countryHouse')),['hasLift']] = 0.0
```

```python
idealista_df.loc[idealista_df['hasLift'].isna(),'propertyType'].value_counts()
```

```python
idealista_df = idealista_df[idealista_df['hasLift'].notna()]
idealista_fe.nan_analysis(idealista_df);
```

Para la prueba de concepto del desafío eliminamos las viviendas con *floor* a nan. Para un proyecto real habría que recurrir a otras fuentes para obtener la planta.

```python
idealista_df = idealista_df[idealista_df['floor'].notna()]
idealista_fe.nan_analysis(idealista_df);
```

```python
idealista_df["floor"].unique()
```

Como en la columna *floor* hay números y texto, vamos a hacer que sólo tome valores numéricos, con estas conversiones:
* *bj* (bajo) a 0
* *ss* (semisótano) a -1
* *st* (sótano) a -1
* *en* (entresuelo) a 0

```python
idealista_df["floor"] = idealista_df['floor'].apply(idealista_fe.floor_str_2_number)
idealista_df["floor"].unique()
```

<a id="otros"></a>
### Otros tratamientos

```python
idealista_df['province'] = idealista_df['province'].apply(lambda x: x.replace('è', 'e'))
idealista_df['municipality'] = idealista_df['municipality'].apply(lambda x: x.replace('è', 'e'))
```

```python
idealista_df['country'] = 'España'
```

```python
idealista_df["hasLift"] = np.where(idealista_df["hasLift"] == 1.0, True, False)
```

```python
idealista_df = idealista_df.astype({'floor': 'int32'})
```

```python
idealista_df
```

```python
idealista_df.info()
```

```python
columnas_str = {}
for index, value in idealista_df.dtypes.items():
    if str(value) == 'object':
        columnas_str[index] = 0

for campo in columnas_str:
    max_length = 0
    for valor in idealista_df[campo].values:
        if max_length < len(valor):
            max_length = len(valor)
    columnas_str[campo] = max_length

display(columnas_str)
```

```python
idealista_df.to_csv(os.path.join(DATA_PATH, 'idealista_alquiler_valencia_clean.csv'), index=False)
```
