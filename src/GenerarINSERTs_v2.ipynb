```python
# Borrar las columnas del DataSet que no se necesitan
cols_borrar = ['city', 'province', 'country']
home_valencia_df.drop(columns = cols_borrar, inplace = True)
```

```python
# CELDA OPCIONAL: Ejecutar mientras no existan estas columnas en el dataset:
# distanceCoworking, safety, score, available
home_valencia_df['distanceCoworking'] = 3000
home_valencia_df['safety'] = np.random.randint(1, 5, home_valencia_df.shape[0], np.int64)
home_valencia_df['score'] = np.random.randint(2, 5, home_valencia_df.shape[0], np.int64)
home_valencia_df['available'] = True
```

```python
home_valencia_df.dtypes
```

```python
idValencia = 1
home_valencia_df['idCity'] = idValencia
```

```python
home_valencia_df
```

```python
for elem in tuple(home_valencia_df.loc[1].values):
    if type(elem) == str:
        print(elem)
```

```python
def create_values(row):
    new_row = []
    for value in row:
        new_value = value
        if type(value) == str:
            new_value = value.replace("'", "''")
        new_row.append(new_value)
    
    new_row = str(tuple(new_row)).replace('"', "'")
    
    return new_row
```

```python
create_values(home_valencia_df.loc[1].values)
```

```python
idealista_valencia_full_df = pd.read_csv(os.path.join(DATA_PATH, 'idealista_alquiler_valencia_full.csv'))
home_valencia_df = pd.merge(home_valencia_df, idealista_valencia_full_df[['propertyCode','thumbnail']], left_on = 'propertyCode', right_on = 'propertyCode')
home_valencia_df.head()
```

```python
display(home_valencia_df.shape)
home_valencia_df.isna().sum()
```

```python
home_valencia_df = home_valencia_df[home_valencia_df['thumbnail'].notna()]
```

```python
home_valencia_df.to_csv(os.path.join(DATA_PATH, 'inserts_home_confoto.csv'), index=False)
```

### Crear fichero SQL INSERTs tabla Home

```python
home_valencia_df = pd.read_csv(os.path.join(DATA_PATH, 'inserts_home_confoto.csv'))
home_valencia_df.isna().sum()
```

```python
colums = ', '.join(home_valencia_df.columns)
with codecs.open(os.path.join(DATA_PATH, 'home_valencia_confoto.sql'), "w", "utf-8") as sql_file:
    for index, row in home_valencia_df.iterrows():
        #sql_file.write("INSERT INTO home (" + colums + ") VALUES " + str(tuple(row.values)) + ";\n")
        sql_file.write("INSERT INTO home (" + colums + ") VALUES " + create_values(row.values) + ";\n")
```
