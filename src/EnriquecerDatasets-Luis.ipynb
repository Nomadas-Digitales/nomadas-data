### Asignar distancia a playa

```python
home_valencia_df = pd.read_csv(os.path.join(DATA_PATH, 'home_valencia.csv'))
home_valencia_df
```

```python
def distance(longitud1, latitud1, longitud2, latitud2):
    '''
    Calcula la distancia en metros entre 2 coordenadas longitud/latitud.

    Parametros:
    * longitud1, latitud1: coordenada 1
    * longitud2, latitud2: coordenada 2
    Resultado:
    * Distancia en metros
    '''
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(latitud1)
    lon1 = radians(longitud1)
    lat2 = radians(latitud2)
    lon2 = radians(longitud2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c * 1000

    return distance
```

```python
home_valencia_df["distanceBeach"] = 0
```

```python
def calcular_dist_a_playa(row):
    #print("longitude:", row['longitude'], "latitude:", row['latitude'])
    distancia = distance(playa_lon, playa_lat, row['longitude'], row['latitude'])
    # print("Distancia a playa:", int(distancia))
    row['distanceBeach'] = int(distancia)
    
    return row
```

```python
playa_lat= 39.483709
playa_lon= -0.324952
```

```python
home_valencia_df = home_valencia_df.apply(calcular_dist_a_playa, axis = 1)
home_valencia_df
```

```python
home_valencia_df.dtypes
```

```python
home_valencia_df.to_csv(os.path.join(DATA_PATH, 'home_valencia.csv'), index=False)
```

## Obtener imágenes de viviendas

```python
idealista_val_df = pd.read_csv(os.path.join(DATA_PATH, 'idealista_alquiler_valencia_full.csv'))
idealista_val_df
```

```python
"https://img3.idealista.com/blur/WEB_LISTING/90/id.pro.es.image.master/a2/30/da/869927479.jpg".rindex(".")
```

```python
def create_photo_name(thumbnail, propertyCode):
    name = str(propertyCode) + thumbnail[thumbnail.rindex("."):]
    
    return name
```

```python
for index, row in idealista_val_df.iterrows():
    thumbnail = row['thumbnail']
    propertyCode = row['propertyCode']
    if str(thumbnail) == 'nan':
        print(propertyCode, "sin foto")
        continue
        
    with open(os.path.join(DATA_PATH, 'photos', create_photo_name(thumbnail, propertyCode)), 'wb') as photo_file:
        response = requests.get(thumbnail, stream=True)

        if not response.ok:
            print("Error al descargar la foto de", propertyCode, response)
            continue

        for block in response.iter_content(1024):
            if not block:
                break

            photo_file.write(block)
```
