```python
#ROBOS

url = 'https://www.lasprovincias.es/valencia-ciudad/barrio-valencia-roba-20170714180957-nt.html'

tablas = pd.read_html(url,decimal=',', thousands='.')
tabla_robos=tablas[0]

tabla_robos

#Ponerle la coma al decimal de % de robos
#Definir tabla de categorías de gravedad
df=tabla_robos

df
```

```python
df.sort_values(by="CP")
```

```python
df2 = df.copy()
```

```python
df2['Barrio'] = df2['Barrio'].apply(lambda x: x.replace(" y ", ", "))
df2
```

```python
new_df=pd.DataFrame()

barrios_con_coma = []

for index, row in df2.iterrows():
    split = row['Barrio'].split(', ')
    #print(row["% de robos"])
    
    nrobarrios = len(split)
    barrios_con_coma.append(row['Barrio'])
    for Barrio in split:
        df_aux=pd.DataFrame({'CP':[row['CP']], 'Barrio':[Barrio.strip()],"% de robos":[row["% de robos"]],"Nro barrios": [nrobarrios]})
        #display(df_aux)
        new_df = new_df.append(df_aux)

new_df = new_df.reset_index(drop=True)
new_df = pd.DataFrame(new_df)

#new_df=new_df.dropna()
new_df=new_df[["CP","Barrio", "% de robos","Nro barrios"]].sort_values(by="CP")

new_df
```

```python
barrios_con_coma
```

```python
new_df2 = new_df.loc[~new_df['Barrio'].isin(barrios_con_coma)].reset_index(drop=True)
new_df2
```

```python
new_df2["% de robos2"]=new_df2["% de robos"]/new_df2["Nro barrios"]
new_df2
```

```python
new_df2["Nro barrios"].value_counts()
```
